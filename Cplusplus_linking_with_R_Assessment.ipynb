## Comparison Table

We rank the psuedocode, Python, C++, and R package from the three assessments in the following table:

| | Replicable | Reproducable | Repeatable | Rerunable | Reusable |
| --- | --- | --- | --- | --- | --- |
| Pseudocode | 1 | 4 | 4 | 1 | 1 |
| Python | 3 | 2(=) | 1(=) | 3(=) | 4 |
| C++ | 4 | 1 | 1(=)  | 3(=) | 2 |
| R Package | 2 | 2(=) | 1(=) | 2 | 3 |

Replicable: The easiest algorithm to replicate is the pseudocode, since all we would need to do is intepret it in a programming language (such as in the other three algorithms). The most difficult algorithm to replicate is the C++ function, since C++ has a more complicated syntax than the other languages. The R package is marginally easier to replicate than the Python due to the documentation.

Reproduceable: The pseudocode could produce different results from the same input if intepreted differently by different people. This means we rank it as the worst of the four options. The C++ algorithm will be easily reproducable, since all C++ code can run in future versions.

Repeatable: There is no random number generation, so all the coded programs are completly deterministic and will produce consistent output for a repeated input. This means we rank them all the same. The exception is the pseudocode, which again could produce different results from the same input if intepreted differently by different people.

Rerunable: The psudeocode algorithm is very precise, and so will be readable as long as the modern English language endures - all we have to do to rerun it is read it again. All three coded functions have lots of comments explaining what the function does. The R package has additional explainations in the documentation, which we can read before rerunning the algorithm. The C++ and Python are less clear, with only a sentance or two explaining what the functions do.

Reusable: Again, we can just reuse the pseudocode by reading it again. The C++ algorithm will be easily reusable, since all C++ code can run in future versions. The R package is marginally easier to reuse than the Python algorithm due to the documentation.

We now explore the three main functions in the package using the following examples: a simple dataset with six points, 1000 points from the standard bivariate normal distribution, and 1000 points on the unit circle. We see that the convex hull plotted in red is clearly the convex hull of the dataset, and that it matches the list of line segments.

```r
test_1 <- list(c(1,1),c(1,2),c(2,1),c(2,2),c(1.5,1.5),c(3,3))
```

```r
jarvis_march(test_1)
```

```r
plot_hull(test_1)
```

```r
set.seed(12345)
test_2 <- list()
for(i in 1:1000){
    test_2[[i]] <- rnorm(2)
}
```

```r
jarvis_march(test_2)
```

```r
plot_hull(test_2)
```

```r
set.seed(12345)
test_3 <- list()
for(i in 1:1000){
    theta <- runif(1,0,pi/2)
    x <- cos(theta)*((2*rbinom(1,1,0.5))-1)
    y <- sin(theta)*((2*rbinom(1,1,0.5))-1)
    test_3[[i]] <- c(x,y)
}
```

```r
jarvis_march(test_3)
```

```r
plot_hull(test_3)
```

We now test the `in_hull` function (which we see from the plots above is accuratly determining whether each point is in the hull or not).

```r
in_hull(test_1,c(1,1))
```

```r
in_hull(test_1,c(1,1.5))
```

```r
in_hull(test_1,c(1.8,1.8))
```

```r
in_hull(test_1,c(4,4))
```

```r
in_hull(test_2,list(c(0,0),c(10,10)))
```

```r
in_hull(test_3,list(c(0,0),c(1,1)))
```
